```python
import matplotlib.pyplot as plt
import pandas as pd

# Données exemple
data = {
    'Étape': ['Revenu initial', 'Ventes supplémentaires', 'Coûts opérationnels', 'Dépenses marketing', 'Nouveaux contrats', 'Autres coûts divers', 'Revenu net'],
    'Valeur': [100000, 50000, -20000, -10000, 30000, -15000, 0]
}

# Convertir en DataFrame
df = pd.DataFrame(data)
df['Position'] = df['Valeur'].cumsum() - df['Valeur']

# Créer le graphique en cascade
plt.bar(df['Étape'], df['Valeur'], bottom=df['Position'], color=['blue', 'green', 'red', 'red', 'green', 'red', 'blue'])
plt.plot(df['Étape'], df['Position'] + df['Valeur'], marker='o', color='black')  # Ligne de cumul

# Personnalisation
plt.title("Graphique en Cascade : Analyse des Revenus")
plt.ylabel("Valeur (€)")
plt.xlabel("Étapes")
plt.xticks(rotation=45, ha="right")
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Afficher
plt.tight_layout()
plt.show()
```
