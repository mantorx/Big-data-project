```python
import pandas as pd 
file_path ="C:/Users/arthu/OneDrive/Bureau/Introduction+python+data+science/pandas"
df = pd.read_csv(file_path + "/covid-19.csv")
data = df.iloc[:,1:]
data.head()
mask = data ["geoId"] == "US"  
us_data = data[mask]
us_data.head()
```

```python
us_data = us_data.reset_index(drop=True)
us_data
```

```python
total_rows = len(us_data)
us_data["growth ratio"] = 0 
for i in range [total_rows -2 ,-1,-1]:
         
    us_data.loc[i, "growth_ratio"] = us_data.loc[ i, "new cases"] / us_data.loc[i+1,"new cases"]
```
